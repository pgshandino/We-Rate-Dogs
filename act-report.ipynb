## Findings

After successfully concluding the gathering, assessing, cleaning and analysis process, the following findings have been made

### 1. Most popular dog breeds



As seen from the chart above, the most popular dog breed present in dataset is the `golden retriever`, appearing more times than any other breed. The `Labrador retriever` and `Pembroke` round up the top three.

### 2. Most liked dog breeds



The `Saluki` breed is the most liked(favouted) dog breed averaging about 24k likes. The ` French bulldog` and `black and tan coonhound` make up number 2 and 3 with 17.9k and 17k respectively. 

### 3. Most retweeted dogs



The most retweeted dog happens to be the `standard poodle` averaging over 10k retweets. The top 3 spots for the mosr retweeted dogs is rounded up by `Afghan hound` and `English springer` averaging about 5.5k and 7k retweets.

### 4. Most popular dog names



As noted from the image above, Cooper is the most popular dog name with 13 appearances. Oliver and Charlie round up the top three with 11 appearance apiece.

## Conclusion

While the goldon retriever appears to be the most popular, it is the saluki breed that appears to be the most liked by users.
